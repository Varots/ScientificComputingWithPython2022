2\. **Outer product**

Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
import numpy as np
u = np.array([1, 3, 5, 7]) ##creating the arrays
v = np.array([2, 4, 6, 8])
print(np.outer(u,v))
outer= np.zeros((u.shape[0],v.shape[0])) ##empty matrix
for i in range(u.shape[0]):
    for j in range(v.shape[0]):
        outer[i][j]= u[i]*v[j]  ##computing the product for each element of the matrix
print(outer)

outer_2=u*v.reshape(len(v),1)     
print(outer_2)
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
import numpy as np
dist= np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
dist_2=dist.reshape(len(dist),1)
m_distances=np.absolute(dist-dist_2) ##taking the absolute of the difference of the two matrixes
m_distances=m_distances*1.609 ## converting km into miles
print("The grid of distances is:\n", m_distances)
````

7\. **Prime numbers sieve**

Compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).

  * Constract a shape (N,) boolean array, which is the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (with `timeit`); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import numpy as np
import timeit
def prime_numbers(N):
    numbers=[x for x in range(N+1)] ## all the numbers
    mask=[] ## empty mask
    for i in range (N+1):
        prime=True
        for x in range (i):
            if(x>1 and i%x==0): ## is this prime?
                prime=False
        mask.append(prime) ##put in the mask the condition
        
    numbers=np.array(numbers)
    prime_numbers=numbers[mask]
    return prime_numbers


print("The prime numbers are :\n",prime_numbers(99))
N=100
start_time=timeit.default_timer()
prime_numbers(N)
end_time=timeit.default_timer()
execution_time=end_time-start_time
print("The execution of the function for N equal to", N, "took", execution_time, "seconds")
N=200
start_time=timeit.default_timer()
prime_numbers(N)
end_time=timeit.default_timer()
execution_time=end_time-start_time
print("The execution of the function for N equal to", N, "took", execution_time, "seconds")
N=400
start_time=timeit.default_timer()
prime_numbers(N)
end_time=timeit.default_timer()
execution_time=end_time-start_time
print("The execution of the function for N equal to", N, "took", execution_time, "seconds") ##computed the time execution for 3 values of n, the execution time is not linear with N

def eratosthenes(N):
    d=np.arange(0,N+1)
    era_mask = np.full(N+1, True)
    for i in range(2, int(N**.5 + 1)):        
        if (era_mask[i] == True): 
            era_mask[i**2::i] = False   ##applying eratosthenes as written in the wiki page
    d = d[era_mask]
    return d

n=99
print("The prime numbers found with the sieve of Eratosthenes are:\n", eratosthenes(n),"\n with N equal to", n)



    
````

8\. **Diffusion using random walk**

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of many random walkers after a given amount of time.

*Hint*: create a 2D array where each row represents a walker, and each column represents a time step.

  * Take 1000 walkers and let them walk for 200 steps
  * Use `randint` to create a 2D array of size $walkers \times steps$ with values -1 or 1
  * Calculate the walking distances for each walker (e.g. by summing the elements in each row)
  * Take the square of the previously-obtained array (element-wise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * **Optional**: plot the average distances ($\sqrt(distance^2)$) as a function of time (step)

````python
import numpy as np
import matplotlib.pyplot as plt
walk=np.random.randint(-1,2,(1000,200)) ##random matrix
print("The random walk matrix is:\n",walk)
distances=walk.sum(axis=1) 
print("The distances for each walker are:\n", distances)
sq_distances=np.square(distances)
print("The square of the distances for each walker are:\n", sq_distances)
mean_steps=distances.mean(axis=0)
print("The mean distances for each time istances are :\n", mean_steps)
##plotting
time_steps=np.arange(1,201) ##vector of 200 time steps
mean_dist=walk.copy()

for i in range(1,200):
    mean_dist[:,i]=mean_dist [:,i]+mean_dist[:,i-1] ## distance for each walker
mean_dist=mean_dist.mean(axis=0)
print("The mean distances for each time step are:\n")
plt.plot(time_steps, mean_dist)
````
